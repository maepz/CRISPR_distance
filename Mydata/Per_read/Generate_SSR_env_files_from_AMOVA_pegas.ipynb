```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('./ALLCRISPRs_haplo_matches_at_5_dist_V6_sorted.txt',sep=',',header=0,index_col=0)
df.drop(10,inplace=True)
df=df.reset_index()
df=df.iloc[:,2:]
df=df.T
df['Sample']=df.index
df=df.sort_index()
df
```

```python
df2=pd.DataFrame(columns=['Sample','read_id'])

for i in df.columns[:-1]:
    d=df.iloc[:,[i,len(df.columns)-1]].dropna()
    d=d[i].apply(lambda x: range(int(x))).explode()
    # d['Sample']=d.index
    dic={'read_id':['_'.join(item) for item in list(zip(list(d.index),[str(i)]*len(d.index),map(str,d.values)))],'Sample':list(d.index.values),str(i):[1]*len(d.index)}
    dic
    newdf=pd.DataFrame.from_dict(dic)
    newdf
    df2=df2.append(newdf,ignore_index=True)
df2['haplo']=df2.read_id.apply(lambda x: int(x.split('_')[-2]))
df2['ind']=df2.read_id.apply(lambda x: int(x.split('_')[-1]))
df2.index=df2.read_id
df2.sort_values(by=['Sample','haplo','ind'], inplace=True)
df2
d=df2[['Sample','read_id','haplo','ind']]
d
```

```python
d.to_csv('ALLCRISPRs_haplo_matches_at_5_dist_V6_sorted_per_read.txt',header=True,index=True,sep='\t')
```

```python
dic_fill={}
for i in range(len(dist)):
    dic_fill[i]=[dist.iloc[i,int(j)] for j in df2.haplo.values]
    
    
print(len(dic_fill))        
```

```python
df3=pd.DataFrame(columns=df2.index,index=df2.index)
```

```python
for i in range(len(df3.index)):
    haplo=int(df3.iloc[i].name.split('_')[-2])
    df3.iloc[i]=dic_fill[haplo]
```

```python
df3.to_csv('./Runs_on_cluster/run2/arrays_in_whole_dataset_v6_noneg_pairwiseDist_per_read.csv',header=True,index=True,sep='\t', chunksize=100000, compression='gzip' , encoding='utf-8')
```
