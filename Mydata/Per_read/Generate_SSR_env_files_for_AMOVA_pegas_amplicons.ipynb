```python
import pandas as pd
import numpy as np
```

```python
# Amplicons
amplicons = pd.read_csv('../Amplicons_haplotypes_all.txt', sep='\t',header=0, index_col=0)
amplicons['Sample'] = [item[12:] for item in amplicons.index.values]
amplicons.index=[item[12:] for item in amplicons.index.values]
amplicons=amplicons.sort_index()
amplicons

locus_dic={'LpxA':['LpxA_CC','LpxA_CT','LpxA_TC','LpxA_TT'],'PleD':['PleD_AA', 'PleD_GA','PleD_GG', 'PleD_AG'],'TufB':['TufB_AA', 'TufB_AC', 'TufB_GA', 'TufB_GC']}

# locus='LpxA'
# locus='PleD'
locus='TufB'


df = amplicons.loc[:,locus_dic[locus]+['Sample']]
df
```

```python
df2=pd.DataFrame(columns=['Sample','read_id'])

for i in range(0,len(df.columns)-1):
    print(i)
    d=df.iloc[:,[i,len(df.columns)-1]].dropna()
#     print(d)
    idx=df.columns[i]
    d=d[idx].apply(lambda x: range(int(x))).explode()
    print(d)
    dic={'read_id':['_'.join(item) for item in list(zip(list(d.index),[str(idx)]*len(d.index),map(str,d.values)))],'Sample':list(d.index.values),str(i):[1]*len(d.index)}
    dic
    newdf=pd.DataFrame.from_dict(dic)
    newdf
    df2=df2.append(newdf,ignore_index=True)
df2['haplo']=df2.read_id.apply(lambda x: str(x.split('_')[-2]))
df2['ind']=df2.read_id.apply(lambda x: int(x.split('_')[-1]))
df2.index=df2.read_id
df2.sort_values(by=['Sample','haplo','ind'], inplace=True)
df2
d=df2[['Sample','read_id','haplo','ind']]

d

# d.to_csv(locus+'_per_read.txt',header=True,index=True,sep='\t')
df2
```
